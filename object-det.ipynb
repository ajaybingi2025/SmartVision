```python
import torch
from PIL import Image
import matplotlib.pyplot as plt
import pandas as pd

# Load your trained model
model = torch.hub.load('ultralytics/yolov5', 'custom', path='runs/train/exp/weights/best.pt')

# Load an image
image_path = '/content/objdet4.jpeg'
image = Image.open(image_path)

# Run inference
results = model(image)

# Set confidence threshold
results.conf = 0.3

# Show results
results.show()  # This will display the image with detected bounding boxes

# Get predictions in pandas format
predictions = results.pandas().xyxy[0]
print(predictions)  # This will print the detected objects

# Count the number of products detected for each brand
brand_counts = predictions['name'].value_counts()
print("\nCount of detected products for each brand:")
print(brand_counts)

# Optional: Display counts visually
brand_counts.plot(kind='bar')
plt.title('Count of Detected Products by Brand')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.show()
```
